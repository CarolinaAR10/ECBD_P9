
# Análisis del Catálogo de Películas y Series de Netflix hasta 2025

Este notebook carga dos archivos separados de películas y series detalladas, los fusiona en un solo archivo combinado y realiza el análisis completo usando el archivo resultante **netflix_dataset_2025.csv**.



## Importar librerías

Se importan las librerías necesarias para la manipulación de datos y visualización.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


## Cargar archivos originales

Leemos los dos archivos CSV originales (películas y series).


```python
movies = pd.read_csv("C:/Users/poxol/Documents/Uni/ECBD/Practica09/DatasetsNuevos/netflix_movies_detailed_up_to_2025.csv")
tv_shows = pd.read_csv("C:/Users/poxol/Documents/Uni/ECBD/Practica09/DatasetsNuevos/netflix_tv_shows_detailed_up_to_2025.csv")

print("Películas:", movies.shape)
print("Series:", tv_shows.shape)
```


## Fusionar datasets

Añadimos la columna 'type' para diferenciar películas y series, luego los combinamos en un solo DataFrame.


```python

movies['type'] = 'Movie'
tv_shows['type'] = 'TV Show'

combined = pd.concat([movies, tv_shows], ignore_index=True)
print("Combinado:", combined.shape)
combined.head()
```


## Guardar el dataset combinado

Guardamos el nuevo archivo fusionado en la misma ubicación con el nombre **netflix_dataset_2025.csv**.


```python

combined.to_csv('netflix_dataset_2025.csv', index=False)
print("Archivo netflix_dataset_2025.csv guardado correctamente.")
```


## Cargar el nuevo archivo fusionado

Leemos el archivo resultante para confirmar su estructura y usarlo en el análisis.


```python

netflix_dataset = pd.read_csv('netflix_dataset_2025.csv')
print("Dimensiones:", netflix_dataset.shape)
netflix_dataset.head()
```


## Inspección de columnas y tipos de datos

Verificamos las columnas disponibles y su tipo para planificar la limpieza.


```python

netflix_dataset.info()
```


## Revisión de valores nulos

Identificamos columnas con valores faltantes.


```python

netflix_dataset.isnull().sum()
```


## Limpieza de datos

Eliminamos filas sin descripción, ya que son necesarias para un sistema de recomendación basado en contenido.


```python

netflix_dataset = netflix_dataset.dropna(subset=['description']).reset_index(drop=True)
print("Dimensiones después de eliminar nulos en descripción:", netflix_dataset.shape)
```


## Visualización: Distribución Películas vs Series

Graficamos la cantidad de películas y series en el catálogo.


```python

plt.figure(figsize=(10,7))
sns.set(style='whitegrid')
ax = sns.countplot(x='type', data=netflix_dataset)
ax.set_title('Distribución de Películas vs Series')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2, height), 
                ha='center', va='bottom', fontsize=12)
plt.show()
```


## Distribución de títulos por año de estreno

Visualizamos cómo se distribuyen los títulos en el tiempo.


```python

plt.figure(figsize=(12,6))
sns.histplot(data=netflix_dataset, x='release_year', bins=30, kde=True)
plt.title('Distribución de Títulos por Año de Estreno')
plt.xlabel('Año de Estreno')
plt.ylabel('Cantidad')
plt.show()
```


## Países con más producciones

Mostramos los principales países productores presentes en el catálogo.


```python

top_countries = netflix_dataset['country'].value_counts().dropna().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Países con más producciones en Netflix')
plt.ylabel('Cantidad')
plt.xlabel('País')
plt.xticks(rotation=45)
plt.show()
```


## Conclusión

En este notebook fusionamos dos datasets detallados (películas y series) en un único archivo **netflix_dataset_2025.csv**, listo para análisis unificado.  
Se realizó exploración, limpieza básica y visualizaciones para entender la distribución del catálogo de Netflix hasta 2025.
